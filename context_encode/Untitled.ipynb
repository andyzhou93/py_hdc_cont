```python
import numpy as np
from scipy.spatial.distance import cdist

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.set_context('poster')
```

```python
N = 8
D = 10000

rim = np.zeros((N,D))
for x in range(N):
    rim[x,:] = np.random.choice([-1.0, 1.0], size=D)


im = np.zeros((N,D))
seed = np.random.choice([-1.0, 1.0], size=D)
flipBits = np.random.permutation(D)
numFlip = np.round(np.linspace(0,5000,N)).astype('int')
for x in range(N):
    im[x,:] = seed
    im[x,flipBits[:numFlip[x]]] *= -1
```

```python
d = cdist(im,im,'hamming')
f,ax = plt.subplots(figsize=(10,8.5))
sns.heatmap(d,vmin=0,vmax=1,ax=ax)
```
