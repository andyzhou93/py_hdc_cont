```python
import numpy as np
from scipy.spatial.distance import cdist
from scipy.stats import binom

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')
sns.set_context('poster')
```

```python
d = np.ones((4,4))*0.5
for n in range(4):
    d[n,n] = 0
    
f,ax = plt.subplots(figsize=(12,10))
sns.heatmap(d,linewidths=1)
# plt.show()
plt.savefig("rim.svg")
```

```python
d = np.array([[0,0.5/3,2*0.5/3,0.5],[0.5/3,0,0.5/3,2*0.5/3],[2*0.5/3,0.5/3,0,0.5/3],[0.5,2*0.5/3,0.5/3,0]])
f,ax = plt.subplots(figsize=(12,10))
sns.heatmap(d,linewidths=1)
# plt.show()
plt.savefig("cim.svg")
```
