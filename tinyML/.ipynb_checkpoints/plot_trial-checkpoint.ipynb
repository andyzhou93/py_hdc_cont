```python
import numpy as np
import h5py
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')
sns.set_context('poster')
```

```python
dataFile = 'data.mat'
file = h5py.File(dataFile,'r')
experimentData = file['experimentData']
keys = list(experimentData.keys())
numTrials, numPositions, numGestures = experimentData[keys[0]].shape
D = file[experimentData[keys[3]][0,0,0]].shape[1]
numEx = 80
```

```python
g = 4
p = 3
t = 0

expLabel = file[experimentData['expGestLabel'][t,p,g]][0,:]
emgFeat = file[experimentData['emgFeat'][t,p,g]][:].T
accFeat = file[experimentData['accFeat'][t,p,g]][:].T

t = np.arange(len(emgFeat))*0.05
```

```python
# chCol = sns.color_palette("ch:s=.25,rot=-.25",64)
chCol = sns.color_palette('plasma',120)
f,ax = plt.subplots(2,1,figsize=(15,10),gridspec_kw={'height_ratios': [2,3]})

for trans in [1,3,5,9,11,13]:
    ax[0].axvline(x=trans, ymin=-1, ymax=1,color='k',linestyle=':')
    ax[1].axvline(x=trans, ymin=0, ymax=15,color='k',linestyle=':')


for n in range(64):
#     ax[1].plot(t,emgFeat[:,n],c=chCol[n])
    ax[1].plot(t,emgFeat[:,n])
ax[0].plot(t,accFeat)

ax[0].set_xlim((0,14))
ax[0].set_ylim((-1,1))
ax[1].set_xlim((0,14))
ax[1].set_ylim((0,0.05))

ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[0].spines['bottom'].set_visible(False)
# ax[0].spines['bottom'].set_position('zero')
ax[0].set_xticklabels([])
ax[0].grid(True,alpha=0.3)
# ax[0].grid(True)
ax[0].set_ylabel('Acceleration (g)')
ax[0].set_axisbelow(True)



ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[1].grid(True,alpha=0.3)
ax[1].set_ylabel('EMG MAV (a.u.)')
ax[1].set_xlabel('Time (s)')
ax[1].set_axisbelow(True)

# plt.show()
plt.savefig("trial.svg")
```
