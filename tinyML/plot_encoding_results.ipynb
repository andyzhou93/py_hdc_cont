```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
sns.set_context('poster')
```

```python
idx = 0

acc = pd.DataFrame(columns=('Accuracy','Test position','Type'))

res = np.load('encoding_dual_stage.npz')
for p in range(8):
    acc.loc[idx] = [res['accSingle'][p,p],'Position %d' % p,'Dual-stage\nclassification']
    idx += 1
    
res = np.load('encoding_none.npz')
for p in range(8):
    acc.loc[idx] = [res['accSuper'][p],'Position %d' % p,'Direct\nsuperposition']
    idx += 1
    
res = np.load('encoding_dual_stage.npz')
for p in range(8):
    acc.loc[idx] = [res['accSuper'][p],'Position %d' % p,'Labeled context\nencoding']
    idx += 1
    
res = np.load('encoding_imu.npz')
for p in range(8):
    acc.loc[idx] = [res['accSuper'][p],'Position %d' % p,'Continuous IMU\nencoding']
    idx += 1

res = np.load('encoding_imu_random.npz')
for p in range(8):
    acc.loc[idx] = [res['accSuper'][p],'Position %d' % p,'Random IMU\nencoding']
    idx += 1
```

```python
f,ax = plt.subplots(figsize=(30,8))
sns.barplot(data=acc,x='Type',y='Accuracy',hue='Test position',ax=ax)
ax.set_xlabel('')
ax.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
ax.set_ylim((0,1))
# plt.show()
plt.savefig('encoding_accuracy.svg')
```
