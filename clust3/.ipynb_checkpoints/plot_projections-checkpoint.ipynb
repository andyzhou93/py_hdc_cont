```python
import pickle
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('./projBase.pickle','rb') as f:
    out = pickle.load(f)
```

```python
dataName = 'allHV.npz'
emgHVType =  'hvRelAcc'

allHV = np.load(dataName)
hv = allHV[emgHVType]
gestLabel = allHV['gestLabel']
posLabel = allHV['posLabel']
trialLabel = allHV['trialLabel']

combGP, groupGP = np.unique(np.column_stack((gestLabel,posLabel)),axis=0,return_inverse=True)

gestures = np.unique(gestLabel)
positions = np.unique(posLabel)
trials = np.unique(trialLabel)

numGestures = len(gestures)
numPositions = len(positions)
numTrials = len(trials)

D = hv.shape[1]
numHV = 80
```

```python
sns.set_context('poster')

f,ax = plt.subplots(figsize=(20,20))

sns.scatterplot(x=out['projAll'][:,0],y=out['projAll'][:,1],hue=['G%d' % g for g in gestLabel],palette='tab20',edgecolor=None,ax=ax,legend=False)
ax.set_title('All positions')
ax.set_xticks([])
ax.set_yticks([])
# ax.legend(loc="lower left", ncol=numGestures,bbox_to_anchor=(1.1, 1.05))
plt.show()

xl = ax.get_xlim()
yl = ax.get_ylim()
```

```python
for p in positions:
    f,ax = plt.subplots(figsize=(20,20))
    sns.scatterplot(x=out['projAll'][:,0],y=out['projAll'][:,1],color=[0.8,0.8,0.8],palette='tab20',edgecolor=None,ax=ax)
    sns.scatterplot(x=out['projAll'][posLabel==p,0],y=out['projAll'][posLabel==p,1],hue=['G%d' % g for i,g in enumerate(gestLabel) if posLabel[i] == p],palette='tab20',edgecolor=None,ax=ax,legend=False)
    ax.set_title('Position %d' % p)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(xl)
    ax.set_ylim(yl)
    plt.show()
```
