```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context('poster')
```

```python
f = open('../clust2/baseline_results.npz','rb')
randRes = np.load(f)
temp = {'Accuracy':randRes['accSuper'], 'Arm Position':np.arange(8), 'Context encoding':'None'}
accData = pd.DataFrame(temp, columns=['Accuracy','Arm Position','Context encoding'])
```

```python
f = open('./baseline_results_random.npz','rb')
randRes = np.load(f)
temp = {'Accuracy':randRes['accSuper'], 'Arm Position':np.arange(8), 'Context encoding':'Random'}
temp = pd.DataFrame(temp, columns=['Accuracy','Arm Position','Context encoding'])
accData = accData.append(temp)
```

```python
f = open('./baseline_results.npz','rb')
randRes = np.load(f)
temp = {'Accuracy':randRes['accSuper'], 'Arm Position':np.arange(8), 'Context encoding':'IMU'}
temp = pd.DataFrame(temp, columns=['Accuracy','Arm Position','Context encoding'])
accData = accData.append(temp)
```

```python
f,ax = plt.subplots(figsize=(30,8))
sns.barplot(data=accData,x='Context encoding',y='Accuracy',hue='Arm Position',ax=ax)
ax.set_ylim((0,1))
plt.legend(bbox_to_anchor=(1.005, 1), loc=2, borderaxespad=0.)
plt.grid(True,which='both')
plt.show()
```
